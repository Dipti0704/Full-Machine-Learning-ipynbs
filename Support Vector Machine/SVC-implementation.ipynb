```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# lets create synthetic datapoints
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_classes=2,n_features=2, n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(pd.DataFrame(X)[0],pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python

from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python

svc.coef_
```

```python
## Prediction

y_pred=svc.predict(X_test)
```

```python

from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
## Prediction

y_pred1=rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(X_train,y_train)
## Prediction
y_pred2=polynomial.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
## Prediction
y_pred3=sigmoid.predict(X_test)
print(classification_report(y_test,y_pred3))

print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```

## ## Support Vector Regression Implementation 

```python
## dataset
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
# independent and dependent features
X = df[['tip', 'sex','smoker','day','time','size']] # independent features
y = df['total_bill'] # dependent feature
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
X_train.head()
```

```python
## feature encoding(Label encoding for sex,smoker,time and one hot encoding for days)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()    
le3 = LabelEncoder()
```

```python
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex']=le1.transform(X_test['sex'])
X_test['smoker']=le2.transform(X_test['smoker'])
X_test['time']=le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
# one hot encoding for days --- columnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'), [3])], remainder='passthrough')
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
X_train=ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
X_test
```

```python
## SVR -- support vector regression
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
# hyperparameter tuning using grid search cv
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print(mean_squared_error(y_test,grid_pred))
print(r2_score(y_test,grid_pred))
print(mean_absolute_error(y_test,grid_pred))
```
